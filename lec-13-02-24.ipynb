```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,6*np.pi,1000)
y=np.cos(x)
```

```python
x # distribution
```

```python
fig,ax=plt.subplots()
ax.plot(x,y) 
ax.set_title("sine wave")
ax.set_ylabel("x- distribution")
ax.set_xlabel(" 4* 3.14 ")
fig.show()
```

```python
fig,ax=plt.subplots()
X=np.linspace(0,10,100)
Y1,Y2=np.sin(X),np.cos(X)
ax.plot(X,Y1,X,Y2)
fig.show()
```

```python
X
```

```python
fig, (ax1,ax2)=plt.subplots(2,1)
ax1.plot(X,Y1,color="C0")
ax2.plot(X,Y2,color="C1")
fig.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3]
y=[2,4,1]
plt.plot(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("MY-first")
plt.show()
```

```python
import matplotlib.pyplot as plt
x1=[1,2,3]
y1=[2,4,1]
plt.plot(x1,y1,label='line1')

x2=[1,3,5]
y2=[2,4,3]
plt.plot(x2,y2,label='line2')

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("MY-second")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[5,4,6,8]
name=["A","B","C","D"]
plt.bar(name,x,color='green')
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[2,4,5,1,2,3]

plt.plot(x,y,color='green',linestyle='dashed',marker='o',markerfacecolor='blue')
plt.xlim(1,8)
plt.ylim(1,8)

plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]


plt.scatter(x,y,color='green',marker='*')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
activities=["apple","google",'microsoft','adani']
slices=[3,4,7,6]
colors=["r","y","g","b"]
plt.pie(slices,labels=activities,colors=colors,startangle=90,shadow=True,explode=(0,0,0.1,0),radius=1.2,autopct='%1.1f%%')
plt.legend()
plt.show()
```
