```python
# write aprogram to remove comment lines from a program
z=open("new.txt",'w')
f=open("sample.txt",'r')
for line in f:
    for i in line:
        if i=='#':
            break
        else:
            z.write(i)
z.close()
f.close()
```

```python
dic={}
f=open("x.txt","r")
for lines in f:
    words=lines.split()
    for word in words:
        if word not in dic:
            dic[word]=1
        else:
            dic[word]+=1
print(dic)
```
